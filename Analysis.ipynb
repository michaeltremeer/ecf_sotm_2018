Drop-shipping has become increasingly popular over the past few years, and I expected drop-shippers to be younger and have less experience than the more traditional manufacturing model. Interestingly, we see that while drop-shippers tend to have less experience than owners of manufacturing businesses, the average age actually tends to be higher. Having said that, drop-shipping does have the greatest spread of owner ages, suggesting that it does have a lower barrier to entry and is easier for anyone to get involved with.

**One important number to take note of is the annual revenues by business model. Our drop-shippers in this survey average 2-2.5 times less revenue than other models, and this will be important to keep in mind when we start to look at some of the other statistics (such as total number of employees/contractors) in the next tables**

Below are two tables which compare all models across all numeric factors. Before, diving in, there are couple things to note:
1. To enable calculation of mean statistics and correlations between variables, many of the responses have been converted from a range to a single number (eg, 5-10 employees becomes 7.5). In most cases I've taken the mid-point of each range, although I've adjusted this where a distribution is heavily skewed in one direction or where there is no upper limit on the final range.
2. Keep in mind that for some of the variables (such as number of SKUs), the distribution of values is very skewed towards the tail (eg a small number of stores with 25,000 SKUs) and this disproportionately increases the average.

Some interesting takeaways:
* The more involved in manufacturing and sourcing the business is, the more experience the owner has (4.7 years for Drop-shipping vs 8.3 years for Hybrid models).
* The less the business model has to do with manufacturing, the more SKUs they sell

Having said that, the average annual income reported by drop-shippers is only $31k per year, compared to $54k-$67k for each of the other models
* The net profit margin of all businesses is between 13-20%
* Private label retailers report the highest percentage their revenue from Amazon sales, with 43% compared to 10-27% for other business models
* Most businesses have 3-6 months of runway in the bank as cash, with more capital intensive models towards the higher end of that range.
